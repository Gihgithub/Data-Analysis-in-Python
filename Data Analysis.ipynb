### Handling Outliers

```python
# Using PERCENTILES
# PERCENTILE of 1 for lower limit and PERCENTILE of 99 for Upper limit
# A percentile divides the data into 100 groups 

upper_limit = df_sales['Revenue'].quantile(0.99)
lower_limit = df_sales['Revenue'].quantile(0.01)
df_sales = df_sales[(df_sales['Revenue'] < upper_limit) & (df_sales['Revenue'] > lower_limit)]
df_sales.head()
```

### Feature Splitting

```python
df_students = pd.read_csv('students.csv')
df_students.head()
```

```python
# Split the name column into first and last name based on a space
df_students['First_Name'] = df_students.Student.str.split(" ").map(lambda var: var[0])
df_students['Lirst_Name'] = df_students.Student.str.split(" ").map(lambda var: var[1])
df_students.head()
```

#### Feature Trasformation

```python
df_students = pd.read_csv('students.csv')
```

```python
# Create performance grade fucntion
def performance_grade(score):
    if score>=75:
        return 'A'
    elif score < 75 and score > 25:
        return 'B'
    else:
        return 'C'

# Apply performance grade function on Whole DataFrame using apply() function
df_students['Grade'] = df_students.Marks.apply(performance_grade)
df_students.head()
```
